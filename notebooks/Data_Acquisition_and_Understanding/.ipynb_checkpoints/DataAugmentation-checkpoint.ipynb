# Data Augmentation for audio

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
import librosa

from sklearn.model_selection import train_test_split
```

```python
os.chdir('../../')
```

```python
# Load preprocessed audio and labels
samples = np.load('data/samples.npy', allow_pickle=True)
labels = np.load('data/labels.npy', allow_pickle=True)
```

### Noise Injection

```python
# Noise injection function
def noise_injection(data, noise_factor):
    noise = np.random.rand(8000)
    augmented_sample = data + noise_factor*noise
    return augmented_sample
```

```python
noisy_samples = noise_injection(samples, 0.03)
```

### Augmented Dataset

```python
plt.rcParams['figure.figsize'] = [10, 6]
plt.style.use('seaborn')

fig, ax = plt.subplots(2,1, sharex=True)
ax[0].plot(samples[10000])
ax[0].set_title('Happy', fontsize=20)
ax[1].plot(noisy_samples[10000])
ax[1].set_title('Noisy Happy', fontsize=20)
fig.tight_layout(pad=1)
```

```python
# Happy
ipd.Audio(samples[10000], rate=8000)
```

```python
# Noisy Happy
ipd.Audio(noisy_samples[10000], rate=8000
         )
```

```python
# Split into throw and keep 
noisyNo, noisyYes, noisylabsNo, noisylabsYes = train_test_split(noisy_samples, 
                                                                labels, 
                                                                test_size=0.3, 
                                                                random_state=420,
                                                                stratify=labels)
```

```python
# Concatenate datasets
aug_samples_3 = np.concatenate((samples, noisyYes))
aug_labels_3 = np.concatenate((labels, noisylabsYes))
```

```python
# Split into Train and Test sets (also shuffle)
X_train_3, X_test_3, d_train_3, d_test_3 = train_test_split(aug_samples_3, 
                                                            aug_labels_3, 
                                                            test_size=0.3, 
                                                            random_state=1992,
                                                            stratify=aug_labels_3)

X_train_3 = np.array(X_train_3)[:,:, np.newaxis]
X_test_3 = np.array(X_test_3)[:,:, np.newaxis]
```

```python
# Print the shape of the resulting arrays
print(X_train_3.shape)
print(X_test_3.shape)
```

```python
# Save the results
np.save('data/X_train_3', X_train_3)
np.save('data/X_test_3', X_test_3)

np.save('data/d_train_3', d_train_3)
np.save('data/d_test_3', d_test_3)
```
