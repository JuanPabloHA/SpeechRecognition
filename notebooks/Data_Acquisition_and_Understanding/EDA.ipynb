# **Exploratory Data Analysis (EDA)**

```python
import os
import numpy as np
import librosa

%matplotlib inline
import matplotlib.pyplot as plt

import IPython.display as ipd
```

```python
# Working directory
os.chdir('../../')
print(os.getcwd())
```

```python
# EDA
audiopath = 'data/audio/'
words = os.listdir(audiopath)
words.remove('.DS_Store')
words.remove('_background_noise_')


print(words)
print(len(words))

size = [len(os.listdir(audiopath+w)) for w in words]
print(size)

print('\n The total number of words recordings in {}'.format(np.sum(size)))
```

```python
# Number of audio files per word category:
plt.figure(figsize=(15,5)) 
plt.style.use('seaborn')
plt.bar(x=words, height=size)
plt.xticks(rotation=90, fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel("Word category", fontsize=15)
plt.ylabel("# of files", fontsize=15)
plt.show()
```

```python
# Load an example
samples, sample_rate = librosa.load(audiopath+'happy/0b77ee66_nohash_1.wav')
```

```python
# Example for the word happy
plt.figure(figsize=(10,5))
plt.plot(samples)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.title("Raw wave of the word happy ({}Hz)".format(sample_rate), fontsize=20)
plt.ylabel("Amplitude", fontsize=15)
plt.show()
```

```python
# Audio sample
ipd.Audio(samples, rate=sample_rate)
```

```python
# Resampling word happy
target_sr = 8000
samples_rs = librosa.resample(samples, sample_rate, target_sr)
```

```python
plt.figure(figsize=(10,5))
plt.plot(samples_rs)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.title("Resampled wave of the word happy ({}Hz)".format(target_sr), fontsize=20)
plt.ylabel("Amplitude", fontsize=15)
plt.show()
```

```python
# Audio sample
ipd.Audio(samples_rs, rate=target_sr)
```
